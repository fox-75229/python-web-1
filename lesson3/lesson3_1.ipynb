```python
# å®šç¾©çˆ¶é¡åˆ¥ Person
class Person:
    def __init__(self, name):  # åˆå§‹åŒ–æ–¹æ³•ï¼Œå»ºç«‹ç‰©ä»¶æ™‚æœƒè‡ªå‹•åŸ·è¡Œ
        self.name = name       # å°‡åå­—å­˜åˆ°å±¬æ€§ name
    
    def greeting(self):        # ä¸€èˆ¬æ–¹æ³•ï¼Œå°å‡ºå•å€™èª
        print(f"æ‚¨å¥½ {self.name}")


# å®šç¾©å­é¡åˆ¥ studentï¼Œç¹¼æ‰¿è‡ª Person
class student(Person):
    def __init__(self, name, chinese, english, math):
        # å‘¼å«çˆ¶é¡åˆ¥(Person)çš„åˆå§‹åŒ–ï¼Œå…ˆæŠŠ name å­˜èµ·ä¾†
        super().__init__(name)
        # å†åŠ ä¸Šä¸‰å€‹å±¬æ€§ï¼šåœ‹æ–‡ã€è‹±æ–‡ã€æ•¸å­¸ (å‰é¢åŠ  __ ä»£è¡¨ç§æœ‰è®Šæ•¸)
        self.__chinese = chinese
        self.__english = english
        self.__math = math
    
    # getter (ç”¨ @property åŒ…è£ï¼Œè®“å‘¼å«ç«¯åƒè®€å±¬æ€§ä¸€æ¨£å–å€¼)
    @property
    def chinese(self):
        return self.__chinese
    
    @property
    def english(self):
        return self.__english

    @property
    def math(self):
        return self.__math

    # è¦†å¯«çˆ¶é¡åˆ¥çš„ greeting()
    def greeting(self):
        super().greeting()   # å…ˆå‘¼å«çˆ¶é¡åˆ¥çš„ greeting()
        print("æ‚¨å¥½:æˆ‘æ˜¯å­¸ç”Ÿ")  # å†åŠ ä¸Šå­¸ç”Ÿè‡ªå·±çš„å•å€™
    
    # è¨ˆç®—ç¸½åˆ†çš„æ–¹æ³•
    def total(self):
        return self.chinese + self.english + self.math
    
    # è¨ˆç®—å¹³å‡çš„æ–¹æ³•
    def average(self):
        return self.total() / 3.0
    
    # ç•¶ print(s1) æ™‚ï¼Œè‡ªå‹•å‘¼å«é€™è£¡
    def __repr__(self):
        return f"""
        æˆ‘æ˜¯ä¸€å€‹å­¸ç”Ÿ
        æˆ‘çš„åå­—æ˜¯{self.name}
        æˆ‘çš„ç¸½åˆ†æ˜¯{self.total()}
        æˆ‘çš„å¹³å‡æ˜¯{self.average()}
        """
```

```python
# å»ºç«‹ä¸€å€‹ student ç‰©ä»¶ï¼Œåå­— kevinï¼Œä¸‰ç§‘åˆ†æ•¸ï¼šåœ‹æ–‡100ã€è‹±æ–‡78ã€æ•¸å­¸92
s1 = student("kevin", english=78, math=92, chinese=100)
# å°å‡ºåå­— (ä¾†è‡ªçˆ¶é¡åˆ¥)
print(s1.name)     
# å–ç”¨åœ‹è‹±æ•¸ä¸‰ç§‘åˆ†æ•¸ (å› ç‚ºæœ‰ç”¨ @propertyï¼Œæ‰€ä»¥çœ‹èµ·ä¾†åƒè®€å±¬æ€§)
print(f"åœ‹æ–‡åˆ†æ•¸:{s1.chinese}")
print(f"è‹±æ–‡åˆ†æ•¸:{s1.english}")
print(f"æ•¸å­¸åˆ†æ•¸:{s1.math}")

# å‘¼å«æ–¹æ³•ç®—ç¸½åˆ†å’Œå¹³å‡
print(f"ç¸½åˆ†:{s1.total()}")
print(f"å¹³å‡:{s1.average()}")

# å‘¼å« greeting()ï¼Œæœƒå…ˆåŸ·è¡Œçˆ¶é¡åˆ¥çš„ â†’ "æ‚¨å¥½ kevin"
# å†åŸ·è¡Œå­é¡åˆ¥çš„ â†’ "æ‚¨å¥½:æˆ‘æ˜¯å­¸ç”Ÿ"
s1.greeting()  
```
